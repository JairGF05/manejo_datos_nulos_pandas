```python
import pandas as pd
import numpy as np
```

```python
diccionario = {
    'Col1': [1,2,3,np.nan],
    'Col2:': [4, np.nan, 6, 7],
    'Col3':['a', 'b', 'c', None]
}
diccionario
```

```python
#Creacion de dataframe con Pandas, al forma se ve m√°s estructurada de esta forma
dataframe_diccionario = pd.DataFrame(diccionario)
dataframe_diccionario
```

```python
#Forma de verificar valores nulos
dataframe_diccionario.isnull()
```

```python
#Verificaci√≥n de nulidad usando un m√©todo n√∫merico
dataframe_diccionario.isnull()*1
```

```python
#Una vez identificados los valores nulos, estos pueden ser reemplazados
dataframe_diccionario.fillna('Missing')
```

```python
#Reemplazar los valosres nulos por la media
dataframe_diccionario.fillna(dataframe_diccionario.mean())
```

```python
#Reemplazar valores, asumiendo la interpolaci√≥n de una serie
dataframe_diccionario.interpolate()
```

```python
dataframe_diccionario
```

```python
#Borrar los valores nulos
dataframe_diccionario.dropna()
```
